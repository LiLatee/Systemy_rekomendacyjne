```python
import numpy as np
import pandas as pd
import json
from sklearn.preprocessing import StandardScaler, normalize
```

```python
train_set_org = pd.read_csv('D:\\Dokumenty\\Systemy_rekomendacyjne\\additional_resources\\2019-master\\data\\train.csv',
                       sep=',',
                      nrows=100,
                      index_col=0)
train_set_org[:3]
```

```python
mask = train_set_org["action_type"] == "clickout item"
df_clicks = train_set_org[mask]
df_item_clicks = (
    df_clicks
    .groupby("reference")
    .size()
    .reset_index(name="n_clicks")
    .transform(lambda x: x.astype(int))
)
df_item_clicks[:5]
```

```python
temp = train_set_org[mask]
temp = temp.astype({'reference':'int'})
merged = temp.merge(
    df_item_clicks[['reference','n_clicks']],
    left_on='reference',
    right_on='reference',
    how='left')
merged[:5]
```

```python
timestamp_difference = merged['timestamp'].max() - merged['timestamp'].min()
timestamp_difference
```

```python
merged2 = merged.copy()

n_buckets = 10
one_step = timestamp_difference/n_buckets
for step in range(n_buckets+1, 1, -1):
    mask = merged2['timestamp'] < merged2['timestamp'].min() + step * one_step
    merged2.loc[mask, 'weight_of_time'] = step * 0.1
merged2[:5]
```

```python
# merged['weight_of_time'] = (merged['timestamp']-merged['timestamp'].min())/timestamp_difference
merged2['weight*n_clicks'] = merged2['weight_of_time']*merged2['n_clicks']
merged2[:5]
```

```python
merged2.to_csv('D:\\Dokumenty\\Systemy_rekomendacyjne\\additional_resources\\2019-master\\data\\train_weight.csv', sep=',', index=False)
```
