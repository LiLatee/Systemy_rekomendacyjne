```python
from sklearn.model_selection import train_test_split
x_train, x_test = train_test_split(train_set, test_size=0.2)
```

```python
x_train.shape
```

```python
temp = False
last_session_id_train_set = x_train.iloc[x_train.shape[0]-1,0]
last_session_id_test_set = x_test.iloc[x_test.shape[0]-1,0]
if last_session_id_test_set.equals(last_session_id_train_set):
    temp = True
    
while temp
    x_train.append(x_test.iloc[x_test.shape[0]-1, :])
    x_test = x_test.iloc[0:x_test.shape[0]-1, :]
    
    last_session_id_train_set = x_train.iloc[x_train.shape[0]-1,0]
    last_session_id_test_set = x_test.iloc[x_test.shape[0]-1,0]
    if last_session_id_test_set.equals(last_session_id_train_set):
        temp = True
    else:
        temp = False
```
